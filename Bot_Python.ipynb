```python
from asyncio.tasks import wait_for
import asyncio
import discord
import random 
import wikipedia #installez la librairie wikipedia

intents = discord.Intents.all()
intents.messages = True
intents.members = True
intents.typing = True
client = discord.Client(command_prefix="!",intents=intents)

wikipedia.set_lang("fr") #permet d'afficher les sommaires wikipédia en français
@client.event
async def on_ready():
  bot_channel = client.get_channel(1044900450450800660)
  await bot_channel.send("Je suis prêt à prendre mes fonctions Mr. le Président")
  general_channel = client.get_channel(1044898984898068562)
  await general_channel.send("Bonjour, je suis Enigmo_Bot le bot lanceur d'énigme, si vous le souhaitez je peux jouer avec vous, pour avoir accès au règles du jeu écrivez 'règles' dans le chat. Je dispose aussi d'un moyen d'afficher les sommaires wikipedia en tapant: 'search' puis plus loin ma recherche avec des points pour remplacer les espaces.")
  print("Le bot est prêt !")

@client.event
async def on_member_join(member):
  general_channel = client.get_channel(1044898984898068562)
  await general_channel.send("Bienvenue, à toi fier(e) partisan(e)  "+member.name)

joueur_en_jeu = []

@client.event
async def on_message(message):
  if message.author == client.user:
    return
  
  message.content = message.content.lower()

  if message.content.startswith("bonjour"):  
   await message.channel.send("Bonjour !")

  def wiki_summary(arg):
    definition = wikipedia.summary(arg,sentences=3,chars=1000,auto_suggest=True,redirect=True)
    return definition

  if message.content.startswith('search'):
    if message.author.id not in joueur_en_jeu:
      words=message.content.split()
      mots_important=words[0:]
      try:
        search= discord.Embed(title="RESULT.....",description=wiki_summary(mots_important), colour=discord.Colour.red())
        await message.channel.send(content=None,embed=search) 
        message = await client.wait_for("message") 

      except:
        searchsummary=wikipedia.suggest(mots_important)
        if searchsummary!=None:
          suggestion=discord.Embed(title="heuu!!",description="je n'es pas retrouver votre description vous pensez a "+ str(searchsummary)+" ??", colour=discord.Colour.brand_red())
          await message.channel.send(content=None, embed=suggestion)
          message = await client.wait_for("message")
          if message.content.lower() in ("oui","yes"):
            search= discord.Embed(title="RESULT.....",description=wiki_summary(searchsummary), colour=discord.Colour.red())
            await message.channel.send(content=None,embed=search) 
          else:
            await message.channel.send("entrer la bonne reponse alors ")
        else:
          error=discord.Embed(title="olalahhh!!",description="je n'es pas retrouver votre description verifier l'orthographe de votre mot ..et verifier si votre requète est de la forme:" +"search jeff.bezos ou search lionnel.messi ", colour=discord.Colour.brand_red())
          await message.channel.send(content=None, embed=error)
          message = await client.wait_for("message")
    else:
      await message.channel.send("Vous ne pouvez pas faire de recherche pendant une chasse au trésor !")
    
  if message.content == "règles" and message.author.id not in joueur_en_jeu: #Si le msg est strictement "règles" le bot repond qwerty
    await message.channel.send("Vous avez demandé les règles de la chasse au trésor, pour commencer à jouer écrivez chasse dans le chat, ensuite vous devrez confirmer. Lorsque le jeu est lancé une série de question vous est posé, vous avez 2 min pour répondre à chacune des énigmes. Vous pouvez entrer 'score' pour connaitre votre score et 'reset' pour relancer la partie")
  
  if message.content == "chasse" and message.author.id not in joueur_en_jeu:
    joueur_en_jeu.append(message.author.id)
    compteur = 0
    Enigme1 = "Je suis né à Londres, ma date de naissance correspond à la date d'une signature d'une lettre du général de Gaulle à son futur ministre des anciens combattant. J'ai étudié la physique à l'université d'oxford durant 3 ans. Qui suis je ?"
    Enigme2 = "Je suis  un inventeur américain, mort à l'âge de 88 ans, j'ai obtenu une médaille John von Neumann, j'étais vivant pendant la seconde guerre mondiale. Je suis, je suis, je suis ?"
    Enigme3 = "Je suis né 5 jour avant la fin de la  Bataille de Grèce, aussi appelée opération Marita, à Amsterdam. Je suis un ingénieur américain ayant été intronisé au  temple de la renommée d'internet. Qui suis je ?"
    Enigme4 = "Je porte un pseudonyme qui me provient d'une marque de céréales, je suis de nationalité américaine, j'ai été condamné à 2 mois de prison. Qui suis je ?"
    Enigme5 = "Je suis un ingénieur français dans le domaine de l'informatique, j'ai fait l'école polytechnique,  j'ai crée un réseau portant le même nom qu'un archipel grec. Qui suis je ?"
    Enigme6 = "Mes travaux fondent la science de l'informatique, j'ai permis en continuant les travaux d'un mathématicien polonais qui ont permis le décryptage de code allemands durant la seconde guerre mondiale. Qui suis je ?"
    Enigme7 = "Je suis l'inventeur d'un langage de programmation, j'ai fait carrière dans le même laboratoire que mon père ou j'ai codévelopper un système d'exploitation, on me désigne parfois par mon adresse électronique dans ce laboratoire. Qui suis je ?"
    Enigme8 = "J'ai cofondé une entreprise avec deux associés dont un est parti au bout de 12 jours, quelques années après avoir crée mon entreprise je suis forcé de démissionner, je fonde donc une nouvelle entreprise qui se fera racheter par mon ancienne entreprise ce qui me permettra de revenir à sa tête. Qui suis je ? "
    Enigme9 = "À 12 ans j’ai créé l’équivalent d’un Wikipédia, j’ai développé un framework Python (web.py), j’ai milité pour une vraie protection juridique des lanceurs d’alerte et la transparence gouvernementale concernant les données officielles. Qui suis je?"
    
    
    
    Enigmes = [Enigme1,Enigme2,Enigme3,Enigme4,Enigme5,Enigme6,Enigme7,Enigme8,Enigme9]
    
    Indices = {
        Enigme1:"J'ai reçu un prix Turing",
        Enigme2:"Ma première invention porte le nom d'un célèbre personnage de Walt Disney",
        Enigme3:"J'ai reçu un prix Prix Princesses des Asturies",
        Enigme4:"J'ai été condamné pour avoir exploité une faille qui me permettais de téléphoner gratuitement",
        Enigme5:"J'ai été fait chevalier de la légion d'honneur puis officier",
        Enigme6:"Une légende ténace dit que ma mort à inspiré le logo d'une célèbre marque d'électronique",
        Enigme7:"J'ai été décoré de la médaille Richard-Hamming",
        Enigme8:"Le World Wide Web a été inventé sur une de mes machines.",
        Enigme9:"JSTOR"
    }

    Reponses = {
        Enigme1:"tim berners-lee",Enigme2:"douglas engelbart",Enigme3:"ray tomlison",Enigme4:"captain crunch",Enigme5:"louis pouzin",Enigme6:"alan turing",Enigme7:"dennis ritchie",Enigme8:"steve jobs",Enigme9:"aaron swartz"
    }

    Images = {
        Enigme1:"https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Tim_Berners-Lee_closeup.jpg/390px-Tim_Berners-Lee_closeup.jpg",
        Enigme2:"https://upload.wikimedia.org/wikipedia/commons/thumb/7/78/Douglas_Engelbart_in_2008.jpg/330px-Douglas_Engelbart_in_2008.jpg",
        Enigme3:"https://upload.wikimedia.org/wikipedia/commons/0/01/Ray_Tomlinson_%28cropped%29.jpg",
        Enigme4:"https://upload.wikimedia.org/wikipedia/commons/thumb/b/b8/JohnDraprBNK.JPG/390px-JohnDraprBNK.JPG",
        Enigme5:"https://upload.wikimedia.org/wikipedia/commons/thumb/2/21/M._Louis_POUZIN_2013.jpg/390px-M._Louis_POUZIN_2013.jpg",
        Enigme6:"https://upload.wikimedia.org/wikipedia/commons/thumb/a/a1/Alan_Turing_Aged_16.jpg/330px-Alan_Turing_Aged_16.jpg",
        Enigme7:"https://upload.wikimedia.org/wikipedia/commons/thumb/2/23/Dennis_Ritchie_2011.jpg/390px-Dennis_Ritchie_2011.jpg",
        Enigme8:"https://upload.wikimedia.org/wikipedia/commons/thumb/f/f5/Steve_Jobs_Headshot_2010-CROP2.jpg/260px-Steve_Jobs_Headshot_2010-CROP2.jpg",
        Enigme9:"https://upload.wikimedia.org/wikipedia/commons/thumb/2/26/Aaron_Swartz_2_at_Boston_Wikipedia_Meetup%2C_2009-08-18.jpg/260px-Aaron_Swartz_2_at_Boston_Wikipedia_Meetup%2C_2009-08-18.jpg"
    }

    player = message.author.id
    channel = message.channel
    Enonce = None
    Score = 0
    ScoreM = 5 * len(Enigmes)
    Point = 5
    def check(message):
     return message.channel == channel and player == message.author.id #fonction qui permet de vérifier si l'utilisateur qui envois le message est le même que celui qui est enregistré dans la variable player et si le channel dans lequel il ecrit est le même que celui du lancement de partie
    await asyncio.sleep(1)
    await message.channel.send("Vous avez entré le mot clé pour commencer une chasse au trésor, souhaitez vous confirmer ? oui ou non")
    message = await client.wait_for("message", check = check) 
    while message.content.lower() != "oui" and  message.content.lower() != "non":
        await message.channel.send("La réponse que vous avez donné n'est pas une réponse attendu")
        message = await client.wait_for("message", check = check) 
    if message.content.lower() == "oui":
        await message.channel.send("La chasse au trésor a débuté, une série d'énigme vous attend.")
        while len(Enigmes) > 0:
          Enonce = random.choice(Enigmes) #variable enonce qui prend une énigme de manière aléatoire à chaque debut de boucle
          await asyncio.sleep(1)
          await message.channel.send(Enonce)
          try: #try/except pour éviter le crash après la fin du temps imparti
            message = await client.wait_for("message",timeout=5,check = check)
            while Reponses[Enonce] not in message.content.lower() and message.content.lower() != "reset":
              if  message.content.lower() == "indice":
                if Point == 5:
                  Point = Point - 2
                  await asyncio.sleep(1)
                  await message.channel.send(Indices[Enonce])
                  message = await client.wait_for("message",check = check)
                else:
                  await asyncio.sleep(1)
                  await message.channel.send(Indices[Enonce])
                  message = await client.wait_for("message",check = check)
              elif message.content.lower() == "score":
                await message.channel.send("Votre score: " +str(Score))
                message = await client.wait_for("message",check = check)
              else: 
                momo= discord.Embed(title="Raté",description="Mauvaise réponse", colour=discord.Colour.red())
                await channel.send(content=None, embed=momo)
                await message.add_reaction("❌")
                message = await client.wait_for("message",check = check)
          except: #si le temps imparti est écoulé l'utilisateur ne gagne aucun point et la question est supprimer
            await asyncio.sleep(1)
            await message.channel.send("Vous avez épuisé votre temps de réponse, vous n'avez gagné aucun points")
            Enigmes.remove(Enonce)
            Point = 5
          if Reponses[Enonce] in message.content.lower():
            await message.channel.send(Images[Enonce])
            momo= discord.Embed(title="BRAVO!!",description="Bonne reponse", colour=discord.Colour.green())
            await channel.send(content=None, embed=momo)
            await message.add_reaction("✅")
            Score = Score + Point
            Enigmes.remove(Enonce)
            Point = 5
          elif message.content.lower() == "reset":
            await message.channel.send("Vous avez choisi de reset la partie")
            Point = 5
            Score = 0
            Enigmes = [Enigme1,Enigme2,Enigme3,Enigme4,Enigme5,Enigme6,Enigme7,Enigme8,Enigme9]
        if Score == ScoreM:
          await message.channel.send("https://i.imgflip.com/3m1z5r.jpg")
          await message.channel.send("Vous avez fini avec le score Maximum Bravo !: "+str(Score))
        elif Score > ScoreM/2:
          await message.channel.send("Vous avez fini avec un score de: "+str(Score))
        elif Score == 0:
          await message.channel.send("https://pbs.twimg.com/media/FGuAOVPXsAUIsk1.jpg")
        else:
          await message.channel.send("https://akm-img-a-in.tosshub.com/indiatoday/images/story/202106/cover4_new_1200x768.jpeg?size=690:388")
          await message.channel.send("Vous n'avez même pas obtenu(e) la moitié des points faites un effort, votre score est de :"+str(Score))
        joueur_en_jeu.remove(message.author.id)
    else:
        await message.channel.send("Vous avez refusez la chasse au trésor...")
  elif message.content == "chasse" and message.author.id in joueur_en_jeu:
    await message.channel.send("Vous avez déja lancer une chasse au trésor, vous ne pouvez pas en lancer une deuxième ne soyez pas gourmand !")
client.run("MTA0NDg5OTYyNzMwMTg3OTgzOQ.GwxFr4.PbZReHNPX3IedsH82Xr9sJcQfdG7ir9hJicU-A")
```
